```python
import pandas as pd
import numpy as np
import math
import warnings
warnings.filterwarnings(action='ignore')
```

```python
def make_df(df_path,img_path):
  df_b = pd.DataFrame([['CARPAL',	'TMCP',	'MMCP',	'LMCP',	'IP',	'MPIP',	'LPIP']])

  gender_li = ['Male','Female']
  for gender in gender_li:    
    if gender == 'Male':
      end = 665
    elif gender == 'Female':
      end = 572

  # 1. 이미지 좌표(라벨링) 불러와서 리스트로 저장
    for i in range(1,end+1):
      df = pd.read_csv(df_path+f"/{gender}/labels/{i}_{gender[0]}.txt",sep=' ',names=['label','x','y','w','h'])
      df.sort_values('label',inplace =True, ignore_index=True)
      df.drop('label',axis=1,inplace=True)
      df['xy'] = df.values.tolist()
  ## 남여 번호순으로 집어넣기
      df_b = df_b.append(df[['xy']].T,ignore_index = True)       
  
  # 2. 데이터프레임 헤더 정리
  header = df_b.iloc[0]
  df = df_b[1:]
  df.rename(columns=header, inplace=True)

  # 3. 남여 번호, 성별, 이미지 패스, BA_mean 삽입
  ## 남여의 BA 1,BA 2값 저장된 파일 각각 불러오기
  m_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/3차프로젝트/1.원본/BA(Male).xlsx')
  f_df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/3차프로젝트/1.원본/BA(Female).xlsx')

  for n in range(1,len(df)+1):
    if n < 666:
      df.loc[n,'No'] = n
      df.loc[n,'gender'] = 0
      df.loc[n,'img_path'] = f'{img_path}/{gender_li[0]}/{n}_{gender_li[0][0]}.jpg'
      df.loc[n,'BA_mean'] = round((m_df['BA 1'][n-1]+m_df['BA 2'][n-1])/2,1)
    else:
      df.loc[n,'No'] = n-665
      df.loc[n,'gender'] = 1
      df.loc[n,'img_path'] = f'{img_path}/{gender_li[1]}/{n-665}_{gender_li[1][0]}.jpg'
      df.loc[n,'BA_mean'] = round((f_df['BA 1'][n-666]+f_df['BA 2'][n-666])/2,1)

  df['No'] = df['No'].astype(int)
  df['gender'] = df['gender'].astype(int)

  # 4. 데이터 프레임 정리
  df.reset_index(drop=True, inplace=True)
  df = df[['No', 'gender', 'BA_mean', 'CARPAL','TMCP','MMCP','LMCP','IP','MPIP','LPIP', 'img_path']]
  return df
```

```python
df = make_df('/content/drive/MyDrive/Colab Notebooks/3차프로젝트/3.라벨링','/content/drive/MyDrive/Colab Notebooks/3차프로젝트/2.전처리')
```
